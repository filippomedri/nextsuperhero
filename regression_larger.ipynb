```python
import pandas as pd
import os
import statsmodels.formula.api as sm
import seaborn as sns
%matplotlib inline
```

```python
model_df = pd.read_csv(os.path.join('data','model_larger'))
```

```python
model_df.rename(columns={'Score': 'X1',
                         'NoUsersRating' : 'X2', 
                         'MetaScore':'X3', 
                         'NoReviews-Users':'X4',
                         'NoReviews-Crit':'X5',
                         'OpeningWE(USA)($)':'X6', 
                         'Gross(USA)($)':'X7',
                         'CumulativeWorldWide Gross($)':'X8', 
                         'NoSeriesNextYear':'Y'}, inplace=True)
```

```python
model_df = model_df[['Y','X1','X2','X3','X4','X5','X6','X7','X8']].copy()
```

```python
model_df.to_csv(os.path.join('data','linear_regression_model_larger'))
```

```python
sns.pairplot(model_df,size = 1.2, aspect=1.5)
```

```python
result = sm.ols(formula="Y ~ X1 + X2 + X3 + X4 + X5 + X6+ X7 + X8", data=model_df).fit()
```

```python
result.params
```

```python
result.summary()
```

```python
result4 = sm.ols(formula="Y ~ X1*X5 + X4*X5 + X2*X3 + X3+X4+X5+X7+X8 ", data=model_df).fit()
```
