```python
import pandas as pd
import numpy as np
import os
import re
```

```python
movies_df = pd.read_csv(os.path.join('data','movies_final'))
```

```python
movies_df.count()
```

```python
movies_df_clean = movies_df[np.isfinite(movies_df['Gross(USA)($)'])]
```

```python
movies_df_clean.count()
```

```python
movies_df_clean.head()
```

```python
movies_df_clean['MetaScore'].fillna(movies_df_clean['MetaScore'].median(),inplace=True)
movies_df_clean.head()
```

```python
movies_df_clean.count()
```

```python
movies_df_clean['OpeningWE(USA)($)'].fillna(movies_df_clean['OpeningWE(USA)($)'].median(),inplace=True) 
```

```python
movies_df_clean.head()
```

```python
movies_df_clean['CumulativeWorldWide Gross($)'].fillna(movies_df_clean['Gross(USA)($)'], inplace=True)
```

```python
movies_df_clean.count()
```

```python
movies_df_clean = movies_df_clean[['Title','Duration(Min)','ReleaseDate','Score','NoUsersRating','MetaScore',
'NoReviews-Users','NoReviews-Crit','OpeningWE(USA)($)','Gross(USA)($)','CumulativeWorldWide Gross($)']]
```

```python
movies_df_clean.head()
```

```python
movies_df_clean.sort_values('ReleaseDate')
```

```python
movies_df_clean.to_csv(os.path.join('data','movies_final_clean'))
```

# TV Series Clean Up

```python
tv_series_df = pd.read_csv(os.path.join('data','tv_series_final'))
```

```python
tv_series_df.head(20)
```

```python
tv_series_df['Year'].unique()
```

```python
tv_series_df['Year'] = tv_series_df['Year'].str.replace('\xa0','')
tv_series_df['Year'] = tv_series_df['Year'].str.replace(r'... See all .\n','')
tv_series_df['Year'] = tv_series_df['Year'].str.replace('Unknown','')
tv_series_df['Year'] = tv_series_df['Year'].str.replace(r'.* See all','')
```

```python
tv_series_df['Year'].unique()
```

```python
tv_series_df.to_csv(os.path.join('data','tv_series_final_first_clean'))
```
