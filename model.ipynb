```python
import pandas as pd
import os
import datetime as dt
```

```python
movies_df = pd.read_csv(os.path.join('data','movies_final_clean'))
```

```python
movies_df.columns
```

```python
movies_df['ReleaseDate'] = pd.to_datetime(movies_df.ReleaseDate)
```

```python
movies_df['Year'] = movies_df.ReleaseDate.dt.year
```

```python
movies_df.head()
```

```python
model_df = movies_df[['Score',
                      'NoUsersRating',
                      'MetaScore',
                      'NoReviews-Users',
                      'NoReviews-Crit',
                      'OpeningWE(USA)($)',
                      'Gross(USA)($)',
                      'CumulativeWorldWide Gross($)',
                      'Year']].copy()
```

```python
model_df.head()
```

```python
tv_series_df = pd.read_csv(os.path.join('data','tv_series_final_clean'))
```

```python
tv_series_df.head()
```

```python
series_per_year_df = tv_series_df.groupby('Year').agg('count').reset_index()
```

```python
series_per_year_df.head()
```

```python
year_list = series_per_year_df['Year'].tolist()
```

```python
count_list = series_per_year_df['Title'].tolist()
```

```python
s_p_y_dict = dict(zip(year_list,count_list))
```

```python
s_p_y_dict
```

```python
model_df = model_df[model_df.Year !=2018]
```

```python
model_df.sort_values('Year',inplace=True)
model_df.tail()
```

```python
model_df['NoSeries'] = model_df.apply(lambda x: s_p_y_dict[x['Year']+1.0], axis=1)
```

```python
model_df.tail()
```

```python
model_df.rename(columns={'NoSeries': 'NoSeriesNextYear'},inplace=True)
```

```python
model_df.to_csv(os.path.join('data','model'))
```
